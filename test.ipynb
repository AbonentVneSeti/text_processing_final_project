```python
# Ячейка 1
from utils import load_config
from models.model_use import load_model, generate_paraphrases

config = load_config("config.yaml")
model = load_model(
    config["model_name"],
    config["model_config"],
    checkpoint_path=config["trainer_config"]["output_dir"]
)
print("Model loaded")
```

```python
# Ячейка 2
sentence = input("Введите предложение: ")
paraphrases = model.generate([sentence])
print("Оригинал:", sentence)
print("Парафраз :", paraphrases[0])
```

```python
# Ячейка 3 (необязательная) – демонстрационные метрики для пары примеров
demo_originals = ["Сегодня хорошая погода.", "Искусственный интеллект меняет мир."]
demo_paraphrases = model.generate(demo_originals)
for orig, para in zip(demo_originals, demo_paraphrases):
    print(f"{orig} -> {para}")
```
